```python
import statsmodels.formula.api as smf
```

```python
x = salary[["Salary"]]
y = salary["YearsExperience"]
```

```python
model1 = smf.ols(formula="Salary~YearsExperience" , data=salary)
```

```python
model1 = model1.fit()
```

```python
y_predictd = model1.predict(salary)
```

```python
yp = pd.DataFrame(y_predictd)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
print("Mean squared value :  ", mean_squared_error(x,yp))
print("r squared value is :  ", r2_score(x,yp))
```

### Logrthm Transformation

```python
salary_copy = salary.copy()
```

```python
salary_copy["Salary"] = np.log(salary_copy["Salary"])
salary_copy["YearsExperience"] = np.log(salary_copy["YearsExperience"])
```

```python
salary_copy
```

```python
log_model = smf.ols(formula="Salary~YearsExperience",data=salary_copy).fit()
```

```python
pred = log_model.predict(salary_copy["YearsExperience"])
```

```python
salary_copy["log_predicted"] = pred
```

```python
salary_copy
```

```python
print("Mean squared value :  ", mean_squared_error(salary_copy["Salary"],pred))
print("r squared value is :  ", r2_score(salary_copy["Salary"],pred))
```

### SQRT Transformation

```python
salarycopy = salary.copy()
```

```python
salarycopy["YearsExperience"] = np.sqrt(salarycopy["YearsExperience"])
salarycopy["YearsExperience"] = np.sqrt(salarycopy["YearsExperience"])                                                 
```

```python
salarycopy
```

```python
sqrt_model = smf.ols(formula="Salary~YearsExperience",data=salarycopy).fit()
```

```python
pred_sqrt = sqrt_model.predict(salarycopy["YearsExperience"])
```

```python
salarycopy["predicted_sqrt"] = pred_sqrt
```

```python
salarycopy
```

```python
print("Mean squared value :  ", mean_squared_error(salarycopy["Salary"],pred_sqrt))
print("r squared value is :  ", r2_score(salarycopy["Salary"],pred_sqrt))
```
