```python
delivery.rename(columns={"Delivery Time" : "DeliveryTime" , "Sorting Time" : "SortingTime"},inplace=True)
```

```python
import statsmodels.formula.api as smf
```

```python
linear_model1= smf.ols(formula= "DeliveryTime~SortingTime",data= delivery).fit()
```

```python
pred = linear_model1.predict(delivery["SortingTime"])
```

```python
delivery["Predicted"] = pred
```

```python
delivery
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
print('Mean square error is: ',mean_squared_error(delivery["DeliveryTime"],pred))
print('R2 score is: ',r2_score(delivery["DeliveryTime"],pred))
```

##  Logorithm Transformation

```python
delivery__copy = delivery.copy()
```

```python
delivery__copy["SortingTime"] = np.log(delivery__copy["SortingTime"])
delivery__copy["DeliveryTime"] = np.log(delivery__copy["DeliveryTime"])
```

```python
log_model = smf.ols(formula="DeliveryTime~SortingTime",data=delivery__copy).fit()
```

```python
pred_ = log_model.predict(delivery__copy["SortingTime"])
```

```python
delivery__copy["predicted_log"] = pred_
```

```python
delivery__copy
```

```python
print('Mean square error is: ',mean_squared_error(delivery__copy["DeliveryTime"],pred_))
print('R2 score is: ',r2_score(delivery__copy["DeliveryTime"],pred_))
```

### SQRT Transformation

```python
delivery_copy = delivery.copy()
```

```python
delivery_copy["DeliveryTime"] = np.sqrt(delivery_copy["DeliveryTime"])
delivery_copy["SortingTime"] = np.sqrt(delivery_copy["SortingTime"])
```

```python
delivery_copy.drop("Predicted",axis=1,inplace=True)
```

```python
delivery_copy.head()
```

```python
sqrt_model = smf.ols(formula="DeliveryTime~SortingTime",data=delivery_copy).fit()
```

```python
pred_sqrt = sqrt_model.predict(delivery_copy["SortingTime"])
```

```python
delivery_copy["predicted_sqrt"] = pred_sqrt
```

```python
delivery_copy
```

```python
print('Mean square error is: ',mean_squared_error(delivery_copy["DeliveryTime"],pred_sqrt))
print('R2 score is: ',r2_score(delivery_copy["DeliveryTime"],pred_sqrt))
```

### Reciprocal Transformation

```python
deliverycopy = delivery.copy()
```

```python
deliverycopy.drop("Predicted",axis=1,inplace=True)
```

```python
deliverycopy["DeliveryTime"] = np.reciprocal(deliverycopy["DeliveryTime"])
deliverycopy["SortingTime"] = 1/delivery["SortingTime"]
```

```python
deliverycopy
```

```python
rec_model = smf.ols(formula="DeliveryTime~SortingTime",data=deliverycopy).fit()
```

```python
pred_rec = rec_model.predict(deliverycopy["SortingTime"])
```

```python
deliverycopy["predicted_rec"]  = pred_rec
```

```python
deliverycopy
```

```python
print('Mean square error is: ',mean_squared_error(deliverycopy["DeliveryTime"],pred_rec))
print('R2 score is: ',r2_score(deliverycopy["DeliveryTime"],pred_rec))
```
